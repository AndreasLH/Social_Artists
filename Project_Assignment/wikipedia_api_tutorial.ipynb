```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml
import wikipedia as wp
import requests
from bs4 import BeautifulSoup
import re
import urllib
```

```python
session = requests.Session()
targetWikiPage = "https://en.wikipedia.org/wiki/50_Cent?oldformat=true"


def extractstring(line,flag1, flag2):
    string = ""
    if flag1 in line: # $ is the flag
        dex1=line.index(flag1)
        subline=line[dex1+len(flag1):-1] #leave out flag (+1) to end of line
        dex2=subline.index(flag2)
        string=subline[0:dex2].strip() #does not include last flag, strip whitespace
    return urllib.parse.unquote_plus(string)


if "wikipedia" in targetWikiPage:
    html = session.post(targetWikiPage)
    bsObj = BeautifulSoup(html.text, "html.parser")

    wikiReferences = bsObj.find_all('', {'class': 'Z3988'})
    wikiReferencesBS = BeautifulSoup(str(wikiReferences), "html.parser")

    for span in wikiReferencesBS.find_all():
        title = span['title']

        print(extractstring(title,flag1='rft.atitle=', flag2='&'))
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
 
subject = 'List_of_music_style'
 
url = 'https://en.wikipedia.org/w/api.php'
params = {
            'action':'parse',
            'prop':'text',
            'format':'json',
            'page':subject,
            'section':0,
            'redirects':''
        }
data = requests.get(url, params=params).json()
data
```

```python
soup = BeautifulSoup(data['parse']['text']['*'],'html.parser')
 
infobox = soup.find('section',{'class': 'article_content ng-scope'})
```

```python
infobox
```

```python
a_tags = infobox.find_all('a', href=True)
 
links = []
for tag in a_tags:
    if not tag.text == '' and 'wiki' in tag['href']:
        links.append({'anchor':tag.text, 'title':tag['title']})
```

```python
list(pd.DataFrame(links).title.values)
```
